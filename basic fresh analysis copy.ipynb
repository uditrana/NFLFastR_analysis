```python
import pandas as pd
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 300)

# YEARS = [2020]
YEARS = list(range(2000, 2021))

final = pd.DataFrame()

#ingest
for year in YEARS:
    data = pd.DataFrame()
    plays = pd.DataFrame()
    print(f"Reading {year}")
    data = data.append(pd.read_csv(f"./data/play_by_play_{year}.csv.gz", compression='gzip', low_memory=False).reset_index(drop=True), sort=True)
    plays = data[['game_id', 'qtr', 'quarter_seconds_remaining']]
    plays = plays.loc[((plays.qtr==4) | (plays.qtr==2)) & (plays.quarter_seconds_remaining<=120)].reset_index(drop=True)
    plays = plays.groupby(['game_id', 'qtr'], as_index=False).max()
    plays['secs_from_2MinWarn'] = 120-plays['quarter_seconds_remaining']
    plays = plays.sort_values("secs_from_2MinWarn", ascending=False).reset_index(drop=True)
    final = final.append(plays, sort=True)
    print(plays.head(10))
final = final.sort_values('secs_from_2MinWarn', ascending=False).reset_index(drop=True)
```
