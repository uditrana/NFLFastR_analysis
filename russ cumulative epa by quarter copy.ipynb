```python
import pandas as pd

YEARS = [2018, 2019]

data = pd.DataFrame()

#ingest
for year in YEARS:
    data = data.append(pd.read_csv(f"data/play_by_play_cleaned_{year}.csv.gz", compression='gzip', low_memory=False).reset_index(drop=True), sort=True)

pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 300)
data.reset_index(drop=True, inplace=True)
```

```python
DOWN_UPPER = 2
NEUTRAL = True
neutral = data.loc[ (data.play_type!="no_play") & (data.qtr<5) &
                       (data.down<=DOWN_UPPER) & (data.half_seconds_remaining>120) &
                                 (data.wp>=.2) & (data.wp<=.8)].reset_index(drop=True)
non_neutral = data.loc[ (data.play_type!="no_play") & (data.qtr<5)].reset_index(drop=True)
neutral['wpa'] = neutral['wpa'].apply(lambda x: x*100)
non_neutral['wpa'] = non_neutral['wpa'].apply(lambda x: x*100)
```

```python
def epa_success(epa):
    return 1 if epa>0 else 0

sea_runs = neutral.loc[ (neutral.play_type=="run") & (neutral.posteam=="SEA")].reset_index(drop=True)
sea_pass = neutral.loc[ (neutral.play_type=="pass") & (neutral.posteam=="SEA")].reset_index(drop=True)

carson = sea_runs.groupby(['posteam', 'season', 'game_id', 'qtr']).agg(
                    play_count = ('play_id', 'count'),
                    avg_epa = ('epa', 'mean'),
                    cum_epa = ('epa', 'sum'),
                    avg_wpa = ('wpa', 'mean'),
                    cum_wpa = ('wpa', 'sum'),
                    epa_succ_rate = ('epa', lambda x: (x.map(epa_success).mean()))).round(2).reset_index().groupby(['posteam', 'season', 'qtr']).agg(
                            play_count = ('play_count', 'sum'),
                            avg_epa = ('avg_epa', 'mean'),
                            cum_epa = ('cum_epa', 'mean'),
                            avg_wpa = ('avg_wpa', 'mean'),
                            cum_wpa = ('cum_wpa', 'mean'),
                            epa_succ_rate = ('epa_succ_rate', 'mean')).round(2)
russ = sea_pass.groupby(['posteam', 'season', 'game_id', 'qtr']).agg(
                    play_count = ('play_id', 'count'),
                    avg_epa = ('epa', 'mean'),
                    cum_epa = ('epa', 'sum'),
                    avg_wpa = ('wpa', 'mean'),
                    cum_wpa = ('wpa', 'sum'),
                    epa_succ_rate = ('epa', lambda x: (x.map(epa_success).mean()))).round(2).reset_index().groupby(['posteam', 'season', 'qtr']).agg(
                            play_count = ('play_count', 'sum'),
                            avg_epa = ('avg_epa', 'mean'),
                            cum_epa = ('cum_epa', 'mean'),
                            avg_wpa = ('avg_wpa', 'mean'),
                            cum_wpa = ('cum_wpa', 'mean'),
                            epa_succ_rate = ('epa_succ_rate', 'mean')).round(2)
neutral_final = russ.join(carson, lsuffix='_pass', rsuffix='_run').reset_index()
neutral_final.to_csv('./saved/Seahawks Quarter Breakdowns/neutral_2018_19_wpa_epa.csv', index=False)
neutral_final
```
